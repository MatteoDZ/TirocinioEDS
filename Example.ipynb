```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from sklearn.cluster import KMeans
```

```python
img = cv2.imread("D:/Download/SettoreCentro/images/ndvi_20210602.png")
plt.imshow(img)
```

```python
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img=img.reshape((img.shape[1]*img.shape[0],3))
```

```python
kmeans=KMeans(n_clusters=4)
s=kmeans.fit(img)
```

```python
labels=kmeans.labels_
print(labels)
labels=list(labels)
```

```python
centroid=kmeans.cluster_centers_
print(centroid)
```

```python
percent=[]
for i in range(len(centroid)):
  j=labels.count(i)
  j=j/(len(labels))
  percent.append(j)
print(percent)
```

```python
plt.pie(percent,colors=np.array(centroid/255),labels=np.arange(len(centroid)))
plt.show()
```

```python
plt.pie(percent,colors=np.array(centroid/255),labels=np.arange(len(centroid)))
plt.show()
```
